```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
x=dataset.iloc[:,1].values
y=dataset.iloc[:,-1].values
x=x.reshape(len(x),1)
y=y.reshape(len(y),1)
```

```python
regrossor=RandomForestRegressor(n_estimators=10,random_state=0)
regrossor.fit(x,y)
```

```python
plt.style.use("fivethirtyeight")
plt.scatter(x,y)
plt.plot(x,regrossor.predict(x),c='red',alpha=0.6)
plt.show()
```
