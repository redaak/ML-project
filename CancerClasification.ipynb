# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
breas_cancer=load_breast_cancer()
```

```python
print(dir(breas_cancer))
```

```python
data=pd.DataFrame(breas_cancer.data,columns=breas_cancer.feature_names)
```

```python
data['target']=breas_cancer.target
```

```python
data.describe()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#SVM
svreger=SVC(kernel='linear',random_state=0)
svreger.fit(x_train,y_train)
svreger.score(x_test,y_test)
accuracy_score(y_test,np.ceil(svreger.predict(x_test)))
```

```python
#LogisticRegression
logreger=LogisticRegression(random_state=0)
logreger.fit(x_train,y_train)
logreger.score(x_test,y_test)
accuracy_score(y_test,np.ceil(logreger.predict(x_test)))
```

```python
#k-nn
k=KNeighborsRegressor(n_neighbors=12,metric='minkowski',p=2)
k.fit(x_train,y_train)
k.score(x_test,y_test)
```

```python
accuracy_score(y_test,np.ceil(k.predict(x_test)))
```
