# Imports

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from statsmodels.api import Logit
from sklearn.model_selection import train_test_split
```

# Loading Datasets and checking for missing values

```python
dataset=pd.read_csv('breastCancer.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.drop(['id','Unnamed: 32'],axis=1)
dataset
```

```python
x=dataset.iloc[:,1:]
y=dataset.iloc[:,0]
```

# **Encoding the target data set** *where M:1 and B:0*

```python
lb=LabelEncoder()
y=lb.fit_transform(y)
print(y)
```

# **spliting and feature scaling**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.fit_transform(x_test)
```

```python
dataset
```

# **Implementing LogisticRegression** 
## * with Statsmodel
## * with sklearn

```python
logisticReg=LogisticRegression()
logisticReg.fit(x_train,y_train)
accuracy_score(y_test,logisticReg.predict(x_test))
```
